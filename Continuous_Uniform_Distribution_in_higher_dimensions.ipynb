```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Cylinder
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
z = np.linspace(0, 1, 100)
theta = np.linspace(0, 2*np.pi, 100)
theta, z = np.meshgrid(theta, z)
r = 1  # radius
x = r * np.cos(theta)
y = r * np.sin(theta)
ax.plot_surface(x, y, z, alpha=0.5)
ax.set_title('Uniform Distribution Cylinder')
plt.show()

# Rectangle
fig, ax = plt.subplots()
rect = plt.Rectangle((0, 0), 2, 1, fc='lightblue', ec='black')
ax.add_patch(rect)
ax.set_xlim(-0.5, 2.5)
ax.set_ylim(-0.5, 1.5)
ax.set_aspect('equal', adjustable='box')
ax.set_title('Uniform Distribution Rectangle')
plt.show()

# Square
fig, ax = plt.subplots()
square = plt.Rectangle((0, 0), 1, 1, fc='lightgreen', ec='black')
ax.add_patch(square)
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(-0.5, 1.5)
ax.set_aspect('equal', adjustable='box')
ax.set_title('Uniform Distribution Square')
plt.show()
```

A uniform distribution doesn't mean that it has to be a rectangle only, it can be any shape, as long as the pmf is same for each value of x, it is uniform

Even if we have broken pieces of cyinder, cube and other objects, as long as their Z axis length is equal they are in uniform as each x has equal probability

```python
# 3D Uniform Distribution (Cube)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate random points in a cube
n_points = 1000
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)

ax.scatter(x, y, z, c='purple', marker='o', alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Uniform Distribution (Cube)')

plt.show()
```

A 3d cube is also a uniform distriubtion if we assume that all points in the cube have same probability

```python
# Generate random points for a 3D uniform distribution in a cube with range [-1, 1]
x = np.random.uniform(-1, 1, n_points)
y = np.random.uniform(-1, 1, n_points)
z = np.random.uniform(-1, 1, n_points)
```

```python
# Calculate squared distance from the origin
squared_distance = x**2 + y**2 + z**2

# Define radius and squared radius
radius = r  # Use the existing variable 'r'
squared_radius = radius**2

# Create a boolean mask for points within the sphere
mask = squared_distance <= squared_radius

# Filter the points
x_sphere = x[mask]
y_sphere = y[mask]
z_sphere = z[mask]
```

```python
# Create a new figure and a 3D subplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Use a scatter plot to visualize the points within the sphere
ax.scatter(x_sphere, y_sphere, z_sphere, c='blue', marker='o', alpha=0.5)

# Label the x, y, and z axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a title to the plot
ax.set_title('3D Uniform Distribution (Sphere)')

# Display the plot
plt.show()
```

A sphere is one of the easiest ways to visualise a uniform distribution object in 3D, since the 4th axis (pmf is not plotted), we can think that all the points lying on the surface of sphere will have same pmf, hence they all are uniformly distributed
